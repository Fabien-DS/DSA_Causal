```python
import numpy as np
import pandas as pd
import scipy.stats as stats

import doubleml as dml

from dowhy import CausalModel

import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')

import graphviz
```

```python
columns=['conversion', 'coupon'] + [f"X{k}" for k in range (15)] + [f"membership_level_{k}" for k in range(3)] + ['Z']
columns
```

```python
causal_graph = """
digraph {
X0 -> conversion;
X1 -> conversion;
X2 -> conversion;
X3 -> conversion;
X4 -> conversion;
X5 -> conversion;
X6 -> conversion;
X7 -> conversion;
X8 -> conversion;
X9 -> conversion;
X10 -> conversion;
X11 -> conversion;
X12 -> conversion;
X13 -> conversion;
X14 -> conversion;
X5 -> coupon;
X6 -> coupon;
X7 -> coupon;
X8 -> coupon;
X9 -> coupon;
X10 -> coupon;
X11 -> coupon;
X12 -> coupon;
X13 -> coupon;
X14 -> coupon;
membership_level_0 -> conversion;
membership_level_1 -> conversion;
membership_level_2 -> conversion;
membership_level_0 -> coupon;
membership_level_1 -> coupon;
membership_level_2 -> coupon;
coupon -> conversion;
coupon -> Z;
conversion -> Z;
}
"""
```

```python
df_model = pd.DataFrame(columns=columns)
df_model
```

```python
# Define a CausalModel object in DoWhy, indicate treatment & outcome
model= CausalModel(data=df_model, # data columns
        graph=causal_graph, # DAG
        treatment='coupon', # cause of interest, X
        outcome='conversion') # outcome, Y
model.view_model(size=(10,30))
```

## Etape 3 : Estimation de l'effet causal

Maintenant que DoWhy nous a fourni u estimateur, nous pouvons mobiliser des méthodes d'estimation statistiques (statistqiue, ML, DL...) pour estier l'effet.

Le fait d'avoir pu déterminer un bon estimateur causal, ne dédoine pas des considérations statistiques à cette étape :
- classes déséquilibrées
- bonne calibration de l'estimateur (prédictiosn conformes)
- optimisation des hyper paramètres

Pour avoir une ase de comparaison, nous allons partir d'une régression logistique

### Etape 3.1 : récupération du jeu de données et réinitialisation du modèle causal

```python
url = 'https://docs.doubleml.org/tutorial/stable/datasets/data/uplift_data.csv'
df = pd.read_csv(url)
```

Il s'agit d'un jeu de donnée simulé, ce qui nous permet (contrairement à un cas réel) de connaitre la valeur réelle de l'effet du traitemant sur chaque unité traitée ("Individual Treatment Effect" ou ite dans le jeu de données

Il est donc possible de calculer l'effet moyen du traitement sur la pouplation sotate ("Average Treatment Effect" ou ATE) :

```python
# True average treatment effect
ATE = df['ite'].mean()
print(ATE)
```

```python
# True average treatment effect on the treated
ATT = df[df['coupon']==1]['ite'].mean()
print(ATT)
```

Il faut réajuster le modèle causal DoWhy à partir des données pour bénéficier de l'estimateur

```python
# Define a CausalModel object in DoWhy, indicate treatment & outcome
model= CausalModel(data=df, # data columns
        graph=causal_graph, # DAG
        treatment='coupon', # cause of interest, X
        outcome='conversion') # outcome, Y
model.view_model(size=(10,30))
```

```python
identified_estimand = model.identify_effect(proceed_when_unidentifiable=True)
print(identified_estimand)
```

### Etape 3.2 Base model

```python
propensity_strat_estimate = model.estimate_effect(identified_estimand,
                                 method_name="backdoor.propensity_score_weighting",
                                 target_units="ate",
                                 method_params={"weighting_scheme":"ips_weight"})

print(propensity_strat_estimate)
```

### EconML

```python
import econml
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LassoCV
from sklearn.ensemble import GradientBoostingClassifier
dml_estimate = model.estimate_effect(identified_estimand,
                                    method_name="backdoor.econml.dml.DML",
                                    method_params={
                                        'init_params': {'model_y':GradientBoostingClassifier(),
                                                        'model_t': GradientBoostingClassifier(),
                                                        'model_final':LassoCV(fit_intercept=False), },
                                        'fit_params': {}
                                     })
print(dml_estimate)
```
