# Introduction à la causalité pour l'actuaire

La première étape de l'analyse consiste à formaliser les hypothèses sur le processus qui a généré les données.

Il est possible de commencer à le formaliser en échangeant avec des experst du domaien sans même avoir besoin d'attendre d'avoir des données

## Etape 0 : réfléchir au problème avant même d'avoir des données

Il est possible de commencer à poser ses hypothèses sur un projet avant même de collecter des données.
Loin d'être annecdotique, cette étape permet d'expliciter des hypothèses qui sans cela resteraient implicites.
Cette étape est particulièrement adaptée pour échanger avec des **experts métiers**, qui détiennent une expertise du domaine.
En explicitant les hypothèses sur les procéssus à l'origine des données, il est possible de générer des discussions fructueuses.

![DAG](../images/CausalDAG.png)

A titre d'exemple, après une série d'interview de découverte d

nous allons considérer l'impact de l'octroi d'une réduction (`coupon`) sur l'achat de services (`conversion`).

Plusieurs variables confondantes (càd étant des ancêtres communs de `coupon`et `conversion`) existent dont :
- 3 niveaux d'abonnement (`membership_level`).
- 10 variables génériques (`X5` à `X14`)

En complément 4 variables génériques  (`X0` à `X4`) sont des ancêtres de `conversion`

Enfin une variable `Z` est un collider (càd un descendant de  `coupon` et `conversion`)

```python
columns=['conversion', 'coupon'] + [f"X{k}" for k in range (15)] + [f"membership_level_{k}" for k in range(3)] + ['Z']
columns
```

```python
causal_graph = """
digraph {
X0 -> conversion;
X1 -> conversion;
X2 -> conversion;
X3 -> conversion;
X4 -> conversion;
X5 -> conversion;
X6 -> conversion;
X7 -> conversion;
X8 -> conversion;
X9 -> conversion;
X10 -> conversion;
X11 -> conversion;
X12 -> conversion;
X13 -> conversion;
X14 -> conversion;
X5 -> coupon;
X6 -> coupon;
X7 -> coupon;
X8 -> coupon;
X9 -> coupon;
X10 -> coupon;
X11 -> coupon;
X12 -> coupon;
X13 -> coupon;
X14 -> coupon;
membership_level_0 -> conversion;
membership_level_1 -> conversion;
membership_level_2 -> conversion;
membership_level_0 -> coupon;
membership_level_1 -> coupon;
membership_level_2 -> coupon;
coupon -> conversion;
coupon -> Z;
conversion -> Z;
}
"""
```

```python
df_model = pd.DataFrame(columns=columns)
df_model
```

```python
# Define a CausalModel object in DoWhy, indicate treatment & outcome
model= CausalModel(data=df_model, # data columns
        graph=causal_graph, # DAG
        treatment='coupon', # cause of interest, X
        outcome='conversion') # outcome, Y
model.view_model(size=(30,30))
```

## Etape 3 : Estimation de l'effet causal

Maintenant que DoWhy nous a fourni u estimateur, nous pouvons mobiliser des méthodes d'estimation statistiques (statistqiue, ML, DL...) pour estier l'effet.

Le fait d'avoir pu déterminer un bon estimateur causal, ne dédoine pas des considérations statistiques à cette étape :
- classes déséquilibrées
- bonne calibration de l'estimateur (prédictiosn conformes)
- optimisation des hyper paramètres

Pour avoir une ase de comparaison, nous allons partir d'une régression logistique

### Etape 3.1 : récupération du jeu de données et réinitialisation du modèle causal

```python
url = 'https://docs.doubleml.org/tutorial/stable/datasets/data/uplift_data.csv'
df = pd.read_csv(url)
```

## Réajustement du modèle DoWhy avec les données

Il faut réajuster le modèle causal DoWhy à partir des données pour bénéficier de l'estimateur

```python
# Define a CausalModel object in DoWhy, indicate treatment & outcome
model= CausalModel(data=df, # data columns
        graph=causal_graph, # DAG
        treatment='coupon', # cause of interest, X
        outcome='conversion') # outcome, Y
model.view_model(size=(30,30))
```

```python
identified_estimand = model.identify_effect(proceed_when_unidentifiable=True)
print(identified_estimand)
```

### Etape 3.2 Base model

```python
linear_estimate = model.estimate_effect(identified_estimand,
                                        method_name="backdoor.linear_regression",
                                       control_value=0,
                                       treatment_value=1)
print(linear_estimate)
```

```python
causal_estimate_ipw = model.estimate_effect(identified_estimand,
                                 method_name="backdoor.propensity_score_weighting",
                                 target_units="ate",
                                 method_params={"weighting_scheme":"ips_weight",
                                                "min_ps_score":0.1,
                                                "max_ps_score":0.9
                                               }
                                )

print(causal_estimate_ipw)
```

```python
print("Causal Estimate is " + str(causal_estimate_ipw.value))
```

```python
causal_estimate_ipw.get_confidence_intervals(0.99)
```

### EconML

```python
%%time
dml_estimate = model.estimate_effect(identified_estimand,
                                    method_name="backdoor.econml.dml.DML",
                                    method_params={
                                        'init_params': {'model_y':GradientBoostingClassifier(),
                                                        'model_t': GradientBoostingClassifier(),
                                                        'model_final':LassoCV(fit_intercept=False), 
                                                        'discrete_outcome':True,
                                                        'discrete_treatment':True,
                                                        'cv':5
                                                       },
                                        'fit_params': {}
                                     })
print(dml_estimate)
```

```python
dml_estimate.get_confidence_intervals(0.95)
```

```python
dml_estimate.value
```

```python
dml_estimate.params
```

```python
import dowhy.gcm as gcm
from dowhy.graph import build_graph_from_str
causal_model = gcm.ProbabilisticCausalModel(build_graph_from_str(causal_graph))
gcm.auto.assign_causal_mechanisms(causal_model, df)
gcm.fit(causal_model, df)
```

```python
gcm.average_causal_effect(causal_model,
                         'conversion',
                         interventions_alternative={'coupon': lambda x: 1},
                         interventions_reference={'coupon': lambda x: 0},
                         num_samples_to_draw=1000000)
```

## Avec le jeu minimal

```python
dml_data = DoubleMLData(df, 
                        y_col='conversion', 
                        d_cols='coupon',
                        x_cols=[f"X{k}" for k in range(5,15)] + [f"membership_level_{k}" for k in range(3)] 
                       )
```

```python
print(dml_data)
```

```python
dml_plr_tree = DoubleMLIRM(dml_data,
                            ml_g = GradientBoostingClassifier(),
                            ml_m = GradientBoostingClassifier(),
                            trimming_threshold = 0.025,
                            n_folds = 3,
                            n_rep = 1
                          )
```

```python
dml_plr_tree.fit(store_predictions=True)
```

```python
print(dml_plr_tree)
```

```python
groups = df[[f"membership_level_{k}" for k in range(3)]]
```

```python
groups.astype(bool)
```

```python
gate = dml_plr_tree.gate(groups.astype(bool))
np.random.seed(42)
print(gate.confint(level=0.90, joint=True))
```

```python
(df
.assign(membership_level=lambda x : x["membership_level_0"]*0 +  x["membership_level_1"]*1 +  x["membership_level_2"]*2)
.groupby(["membership_level"])["ite"].mean()
)
```
